```python
water['Year'] = water['Year'].apply(lambda yr: yr[4::])
```

```python
state_water_contam = pd.DataFrame({'Occurances': water.groupby(['State Name','Contaminants']).size()}).reset_index()
state_water_contam
```

```python
state_water_contam_norm = pd.DataFrame({'Prevelance': state_water_contam.groupby(['State Name']).transform(lambda x: (x - x.mean()) / x.std())})
state_water_contam_norm
```
